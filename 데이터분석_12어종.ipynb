```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
import os
import matplotlib.pyplot as plt
```

데이터 불러오기(수산물_통합전처리_3컬럼.csv) 

```python
file_path = './data/수산물_통합전처리_3컬럼.csv'

try:
    # 파일 불러오기
    data_all = pd.read_csv(file_path, encoding='utf-8')
    
    print("✓ 파일 로드 성공!")
    print(f"\n데이터 shape: {data_all.shape}")
    print(f"행 수: {len(data_all):,}행")
    print(f"컬럼 수: {len(data_all.columns)}개")
    
    print(f"\n컬럼 목록:")
    for i, col in enumerate(data_all.columns, 1):
        print(f"  {i}. {col}")
    
    print(f"\n처음 5행:")
    print(data_all.head())
    
    print(f"\n데이터 타입:")
    print(data_all.dtypes)
    
    print(f"\n기본 통계:")
    print(data_all.describe())
    
    print(f"\n결측치 확인:")
    print(data_all.isnull().sum())
    
except UnicodeDecodeError:
    print("utf-8 인코딩 실패. cp949로 재시도 중...")
    data_all = pd.read_csv(file_path, encoding='cp949')
    print("✓ cp949 인코딩으로 로드 성공!")
    print(data_all.head())
    
except Exception as e:
    print(f"✗ 오류 발생: {e}")
```

```python
data_all
```

컬럼 순서 재배치 

```python
columns = ['date', 'year', 'month'] + [col for col in data_all.columns if col not in ['date', 'year', 'month']]
data_all = data_all[columns]

print("✓ 컬럼 순서 변경 완료!")
print(f"\n변경된 컬럼 순서:")
for i, col in enumerate(data_all.columns, 1):
    print(f"  {i}. {col}")

print(f"\n처음 5행:")
print(data_all.head())
```

```python
data_all.head(2)
```

 평균가와 낙찰고가,낙찰고가 컬럼의 문자 -> 점수 변환(콤마제거)

```python
for col in ['평균가', '낙찰고가', '낙찰저가']:
    data_all[col] = pd.to_numeric(
        data_all[col].astype(str).str.replace(',', ''), 
        errors='coerce'  # 변환 실패시 NaN으로 처리
    )
    print(f"✓ {col} 변환 완료")
```

전체 데이터에서 어종공통 별  어종의 종류가 몇가지인지 확인

```python
# 공통어종별 어종 분석
summary = data_all.groupby('공통어종').agg({
    '어종': ['nunique', lambda x: ', '.join(x.unique())]
}).reset_index()

summary.columns = ['공통어종', '어종_개수', '어종_목록']
summary = summary.sort_values('어종_개수', ascending=False)

print("공통어종별 어종 종류:")
print(summary.to_string(index=False))

# 각 공통어종별 데이터 개수도 함께
detailed = data_all.groupby('공통어종').agg({
    '어종': ['nunique', 'count']
}).reset_index()
detailed.columns = ['공동어종', '어종_종류수', '전체_데이터수']
detailed = detailed.sort_values('어종_종류수', ascending=False)

print("\n" + "=" * 60)
print("공통어종별 상세 정보:")
print(detailed.to_string(index=False))
```

#### 연도별 어종별

```python
# 연도별 어종별 평균 시세
yearly_avg = data_all.groupby(['year', '어종'])['평균가'].mean().reset_index()
yearly_avg['평균가'] = yearly_avg['평균가'].round(0).astype(int)

print("=" * 60)
print("연도별 어종별 평균 시세:")
print("=" * 60)
print(yearly_avg)

# 피벗 테이블로 보기 좋게
pivot_table = data_all.pivot_table(
    values='평균가', 
    index='어종', 
    columns='year', 
    aggfunc='mean'
).round(0)
```

```python
# 연도별 어종별 평균 시세 데이터프레임
yearly_avg_price = data_all.groupby(['year', '어종'])['평균가'].mean().reset_index()
yearly_avg_price['평균가'] = yearly_avg_price['평균가'].round(0).astype(int)
yearly_avg_price = yearly_avg_price.sort_values(['year', '평균가'], ascending=[True, False])

print("=" * 60)
print("연도별 어종별 평균 시세 데이터프레임:")
print("=" * 60)
print(yearly_avg_price)

print(f"\n총 {len(yearly_avg_price)}개 행")
print(f"연도 범위: {yearly_avg_price['year'].min()} ~ {yearly_avg_price['year'].max()}")
print(f"어종 종류: {yearly_avg_price['어종'].nunique()}개")

# 피벗 테이블로 보기 (어종을 행, 연도를 열로)
print("\n" + "=" * 60)
print("피벗 테이블 형태:")
print("=" * 60)
pivot_df = yearly_avg_price.pivot(index='어종', columns='year', values='평균가')
pivot_df = pivot_df.fillna(0).astype(int)  # NaN을 0으로 채우고 정수로 변환
print(pivot_df)
```

## 규격

#### 어종별 가장 많이 거래되는 규격

```python
data_all['규격'].unique()
```

```python
data_all.head(2)
```

```python
# 1. 어종/규격 조합별 거래 건수 집계
counts = data_all.groupby(['어종', '규격_정제']).size().reset_index(name='count')

# 2. 어종별로 거래량이 가장 많은 규격만 추출
idx = counts.groupby('어종')['count'].idxmax()
most_traded_specs = counts.loc[idx].reset_index(drop=True)

# 3. 집계 건수가 많은 순서대로 내림차순 정렬
most_traded_specs_sorted = most_traded_specs.sort_values('count', ascending=False).reset_index(drop=True)

print("어종별 가장 많이 거래된 규격 (집계 건수순):")
print(most_traded_specs_sorted)
```

```python
most_traded_specs_sorted
```

#### 공통어종별 가장 적게 거래되는 

```python
# 2. 어종별로 거래량이 가장 적은 규격만 추출
idx_min = counts.groupby('어종')['count'].idxmin()
least_traded_specs = counts.loc[idx_min].reset_index(drop=True)

# 3. 집계 건수가 적은 순서대로 내림차순 정렬 (원하면 오름차순)
least_traded_specs_sorted = least_traded_specs.sort_values('count', ascending=True).reset_index(drop=True)

print("어종별 가장 적게 거래된 규격 (집계 건수순):")
print(least_traded_specs_sorted)
```
