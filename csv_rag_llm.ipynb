```python
import numpy as np

def parse_size(value):
    if not isinstance(value, str):
        return np.nan
    # '미' 포함: 숫자 처리
    if '미' in value:
        parts = value.replace('미', '').split('/')
        nums = [float(p) for p in parts if p.isdigit()]
        if len(nums) == 1:
            return nums[0]
        elif len(nums) == 2:
            return np.mean(nums)
        else:
            return np.nan
    # '소', '중소', '중', '대', '특대' 등 포함: 그대로 반환
    for kw in ['소', '중소', '중', '대', '특대', '5코', '3코', '2코', '10코', '4코', '6코']:
        if kw == value:
            return kw
    return np.nan

merged_df['규격_파생'] = merged_df['규격'].apply(parse_size)
```

```python
mm_list = ['110/160', '200/300', '100/150', '350/450', '150/200']
keyword_list = ['소', '중소', '중', '대', '특대', '5코', '3코', '2코', '10코', '4코', '6코']

def parse_size(value):
    if not isinstance(value, str):
        return np.nan
    # mm 단위: 미가 없고 지정된 리스트에 있으면 평균(cm) + 단위(문자)
    if (value in mm_list) and ('미' not in value):
        nums = [float(x) for x in value.split('/')]
        result_cm = np.mean(nums) / 10
        return f"{result_cm}cm"
    # '미' 포함: 평균값 그대로(숫자)
    if '미' in value:
        parts = value.replace('미', '').split('/')
        nums = [float(p) for p in parts if p.isdigit()]
        if nums:
            return np.mean(nums)
        else:
            return np.nan
    # 키워드 그대로
    if value in keyword_list:
        return value
    return np.nan

merged_df['규격_파생'] = merged_df['규격'].apply(parse_size)
```

```python
import os
from pathlib import Path
import pandas as pd
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# 0) 설정
CSV_PATH = "./data/rag_df2.csv"
CHROMA_DIR = "./csv_storage/chroma_fish2"
EMBED_MODEL = "text-embedding-004"
LLM_MODEL = "gemini-2.5-flash"

os.environ["GOOGLE_API_KEY"] = "AIzaSyC_Jj_qbteCJCC5mwCY3AUR05oFCUGGQiQ"

# 1) CSV로더 & 청크생성
def load_and_split(csv_path: str):
    df = pd.read_csv(csv_path)
    chunks = []
    for idx, row in df.iterrows():
        text = (
            f"파일어종: {row['파일어종']}, 전처리: {row['전처리']}, 산지: {row['산지']}, 날짜: {row['date']}, "
            f"기온평균: {row['기온 평균']}, 수온평균: {row['수온 평균']}, 풍속평균: {row['풍속 평균']}, year: {row['year']}, month: {row['month']}, 평균가: {row['평균가']}"
        )
        chunks.append(Document(page_content=text, metadata={"row": idx}))
    return df, chunks

# 2) 임베딩 & 벡터DB
def build_or_load_vectorstore(chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model=EMBED_MODEL)
    if Path(CHROMA_DIR).exists():
        vectordb = Chroma(persist_directory=CHROMA_DIR, embedding_function=embeddings)
    else:
        vectordb = Chroma.from_documents(
            documents=chunks,
            embedding=embeddings,
            persist_directory=CHROMA_DIR,
        )
    return vectordb

# 3) RAG 체인
def make_rag_chain(vectordb):
    retriever = vectordb.as_retriever(search_kwargs={"k": 5})
    system_prompt = """
당신은 한국어로 답하는 수산물 데이터 전문 어시스턴트입니다.
다음 규칙을 반드시 지키세요:
- 아래에 제공된 데이터(row 및 요약)를 근거로 하여 간결하고 정확하게 답변하세요.
- 표, 수치, 날짜, 어종명, 산지명 등은 반드시 원문 데이터를 바탕으로 인용하세요.
- 최근 트렌드, 패턴, 특이사항 설명 또는 비교가 필요한 경우 데이터 중심적으로 서술하세요.
- 요청 내용과 직접 관련된 데이터가 없는 경우, 무리해서 가정하지 말고 '검색된 데이터에 근거한 답변이 어렵다'고 안내하세요.
- 답변 마지막 줄에는 반드시 '출처: 수산물 CSV 데이터'라고 표시하세요.
"""
    prompt = ChatPromptTemplate.from_messages(
        [
            ("system", system_prompt),
            ("human", "질문: {question}\n\n다음은 검색된 데이터 조각입니다:\n{context}"),
        ]
    )
    llm = ChatGoogleGenerativeAI(
        model=LLM_MODEL,
        temperature=0.2,
    )
    def format_docs(docs):
        out = []
        for i, d in enumerate(docs, 1):
            meta = d.metadata or {}
            row = meta.get("row", "N/A")
            out.append(f"[{i}] (row {row}) {d.page_content[:800]}")
        return "\n\n".join(out)
    retriever_step = {
        "context": lambda x: format_docs(retriever.invoke(x["question"])),
        "question": lambda x: x["question"]
    }
    chain = retriever_step | prompt | llm | StrOutputParser()
    return chain

# 5) 머신러닝 모델 학습 (RandomForest + ColumnTransformer)
def train_model(df):
    feature_cols = ['파일어종', '전처리', '산지', '기온 평균', '수온 평균', '풍속 평균', 'year', 'month']
    X = df[feature_cols].copy()
    y = df['평균가']
    # 카테고리형 → str
    for col in ['파일어종', '전처리', '산지', 'year', 'month']:
        X[col] = X[col].astype(str)
    ct = ColumnTransformer([
        ('onehot', OneHotEncoder(handle_unknown="ignore"), [0,1,2,6,7]),
        ('scaler', StandardScaler(), [3,4,5])
    ])
    X_ct = ct.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_ct, y, test_size=0.2, random_state=29)
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    return model, ct, feature_cols

def predict_price(model, ct, feature_cols, input_dict):
    import pandas as pd
    # 새로운 카테고리일 수도 있으니 모두 문자열로 변환
    values = [str(input_dict.get(col)) if col in ['파일어종','전처리','산지','year','month'] else input_dict.get(col) for col in feature_cols]
    X_query = pd.DataFrame([values], columns=feature_cols)
    X_query_ct = ct.transform(X_query)
    pred = model.predict(X_query_ct)[0]
    return pred

# 4) 인덱싱 & 질의 + ML 예측
def main():
    df, splits = load_and_split(CSV_PATH)
    vectordb = build_or_load_vectorstore(splits)
    chain = make_rag_chain(vectordb)
    model, ct, feature_cols = train_model(df)

    questions = [
        "수온이 25도 이상인 날의 어종과 산지를 알려줘.",
        "파일어종이 갈치, 산지는 제주도, 전처리는 냉, 기온 평균 18, 수온 평균 26, 풍속 평균 3, 2025년 1월인 평균가를 예측해줘."
    ]
    for q in questions:
        print("=" * 80)
        print("Q:", q)
        print("-" * 80)
        answer = chain.invoke({"question": q})
        print(answer)
        print("\n")

        if "예측" in q:
            ml_input = {
                '파일어종': '갈치',
                '전처리': '냉',
                '산지': '제주도',
                '기온 평균': 18,
                '수온 평균': 26,
                '풍속 평균': 3,
                'year': 2025,
                'month': 1
            }
            price_pred = predict_price(model, ct, feature_cols, ml_input)
            print(f"ML 모델 경락가 예측: {price_pred:.2f} 원")
            print("\n")

if __name__ == "__main__":
    main()
```

```python
import os
from pathlib import Path
import pandas as pd
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 0) 설정
CSV_PATH = "./data/rag_df2.csv"
CHROMA_DIR = "./csv_storage/chroma_fish3"
EMBED_MODEL = "text-embedding-004"
LLM_MODEL = "gemini-2.5-flash"

os.environ["GOOGLE_API_KEY"] = "AIzaSyC_Jj_qbteCJCC5mwCY3AUR05oFCUGGQiQ"

# 1) CSV로더 & 청크생성
def load_and_split(csv_path: str):
    df = pd.read_csv(csv_path)
    chunks = []
    for idx, row in df.iterrows():
        text = (
            f"파일어종: {row['파일어종']}, 전처리: {row['전처리']}, 산지: {row['산지']}, 날짜: {row['date']}, "
            f"기온평균: {row['기온 평균']}, 수온평균: {row['수온 평균']}, 풍속평균: {row['풍속 평균']}, year: {row['year']}, month: {row['month']}, 평균가: {row['평균가']}"
        )
        chunks.append(Document(page_content=text, metadata={"row": idx}))
    return df, chunks

# 2) 임베딩 & 벡터DB
def build_or_load_vectorstore(chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model=EMBED_MODEL)
    if Path(CHROMA_DIR).exists():
        vectordb = Chroma(persist_directory=CHROMA_DIR, embedding_function=embeddings)
    else:
        vectordb = Chroma.from_documents(
            documents=chunks,
            embedding=embeddings,
            persist_directory=CHROMA_DIR,
        )
    return vectordb

# 3) RAG 체인
def make_rag_chain(vectordb):
    retriever = vectordb.as_retriever(search_kwargs={"k": 5})
    system_prompt = """
당신은 한국어로 답하는 수산물 데이터 전문 어시스턴트입니다.
다음 규칙을 반드시 지키세요:
- 아래에 제공된 데이터(row 및 요약)를 근거로 하여 간결하고 정확하게 답변하세요.
- 표, 수치, 날짜, 어종명, 산지명 등은 반드시 원문 데이터를 바탕으로 인용하세요.
- 최근 트렌드, 패턴, 특이사항 설명 또는 비교가 필요한 경우 데이터 중심적으로 서술하세요.
- 요청 내용과 직접 관련된 데이터가 없는 경우, 무리해서 가정하지 말고 '검색된 데이터에 근거한 답변이 어렵다'고 안내하세요.
- 답변 마지막 줄에는 반드시 '출처: 수산물 CSV 데이터'라고 표시하세요.
"""
    prompt = ChatPromptTemplate.from_messages(
        [
            ("system", system_prompt),
            ("human", "질문: {question}\n\n다음은 검색된 데이터 조각입니다:\n{context}"),
        ]
    )
    llm = ChatGoogleGenerativeAI(
        model=LLM_MODEL,
        temperature=0.2,
    )
    def format_docs(docs):
        out = []
        for i, d in enumerate(docs, 1):
            meta = d.metadata or {}
            row = meta.get("row", "N/A")
            out.append(f"[{i}] (row {row}) {d.page_content[:800]}")
        return "\n\n".join(out)
    retriever_step = {
        "context": lambda x: format_docs(retriever.invoke(x["question"])),
        "question": lambda x: x["question"]
    }
    chain = retriever_step | prompt | llm | StrOutputParser()
    return chain

# 4) 부분 feature 자동 채움 (평균/최빈값 보완)
def complete_features(df, partial_input):
    # partial_input: {'파일어종':..., '산지':..., 'year':..., 'month':...}
    filt = (
        (df['파일어종'] == partial_input['파일어종']) &
        (df['산지'] == partial_input['산지']) &
        (df['year'].astype(str) == str(partial_input['year'])) &
        (df['month'].astype(str) == str(partial_input['month']))
    )
    df_sub = df[filt]
    result = dict(partial_input)  # 우선 입력값 복사
    fill_cols = ['전처리', '기온 평균', '수온 평균', '풍속 평균']
    for col in fill_cols:
        # 범주형은 mode(), 수치형은 mean()
        if len(df_sub) > 0:
            # 해당 조건 subset에 값이 있으면 그것으로 보완
            result[col] = df_sub[col].mode()[0] if df_sub[col].dtype == 'O' else float(df_sub[col].mean())
        else:
            # 전체 평균/최빈값으로 채움
            result[col] = df[col].mode()[0] if df[col].dtype == 'O' else float(df[col].mean())
    return result

# 5) 머신러닝 모델(RandomForest + ColumnTransformer)
def train_model(df):
    feature_cols = ['파일어종', '전처리', '산지', '기온 평균', '수온 평균', '풍속 평균', 'year', 'month']
    X = df[feature_cols].copy()
    y = df['평균가']
    # 카테고리형 → str
    for col in ['파일어종', '전처리', '산지', 'year', 'month']:
        X[col] = X[col].astype(str)
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.preprocessing import OneHotEncoder, StandardScaler
    from sklearn.compose import ColumnTransformer
    from sklearn.model_selection import train_test_split
    ct = ColumnTransformer([
        ('onehot', OneHotEncoder(handle_unknown="ignore"), [0,1,2,6,7]),
        ('scaler', StandardScaler(), [3,4,5])
    ])
    X_ct = ct.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_ct, y, test_size=0.2, random_state=29)
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    return model, ct, feature_cols

def predict_price(model, ct, feature_cols, input_dict):
    import pandas as pd
    values = [str(input_dict.get(col)) if col in ['파일어종','전처리','산지','year','month'] else input_dict.get(col) for col in feature_cols]
    X_query = pd.DataFrame([values], columns=feature_cols)
    X_query_ct = ct.transform(X_query)
    pred = model.predict(X_query_ct)[0]
    return pred

# 6) 인덱싱 & 질의 + ML 예측
def main():
    df, splits = load_and_split(CSV_PATH)
    vectordb = build_or_load_vectorstore(splits)
    chain = make_rag_chain(vectordb)
    model, ct, feature_cols = train_model(df)

    questions = [
        "수온이 25도 이상인 날의 어종과 산지를 알려줘.",
        "2025년 1월 제주도산 갈치 평균가를 예측해줘."
    ]
    for q in questions:
        print("=" * 80)
        print("Q:", q)
        print("-" * 80)
        answer = chain.invoke({"question": q})
        print(answer)
        print("\n")
        # 예측 파트: 간략 입력 정보만 있으면 자동 보완 후 ML 예측
        if "예측" in q:
            # 자연어 파싱 예시(실전은 regex or NLP parsing 적용)
            # 예시 - 실제 질의에서 정보 파싱: 여기선 하드코딩
            partial_input = {
                '파일어종': '갈치',
                '산지': '제주도',
                'year': 2025,
                'month': 1
            }
            # 자동 feature 보완
            ml_input = complete_features(df, partial_input)
            price_pred = predict_price(model, ct, feature_cols, ml_input)
            print(f"간략 정보 기반 ML 평균가 예측: {price_pred:.2f} 원")
            print("\n")

if __name__ == "__main__":
    main()
```
